```python
# 현재 노트북이 있는 폴더 안의 모든 jpeg 파일을 img_001.jpg ~ img_040.jpg 로 변경
from pathlib import Path

# 현재 작업 디렉토리 = 이 .ipynb 파일이 있는 바로 그 폴더
folder = Path.cwd()

print(f"작업 폴더: {folder}\n")

# .jpeg, .JPEG, .jpg, .JPG 모두 잡아오기 (혹시 섞여 있을 수도 있으니까)
files = sorted(folder.glob("*.jpeg")) + sorted(folder.glob("*.JPEG"))

if len(files) == 0:
    print("jpeg 파일이 하나도 없어요. 확장자를 확인해 보세요!")
else:
    print(f"발견된 jpeg 파일 {len(files)}개\n")

# 이름 변경 시작
for i, old_path in enumerate(files, start=1):
    new_name = f"img_{i:03d}.jpg"        # img_001.jpg, img_002.jpg ...
    new_path = folder / new_name
    
    # 같은 이름 있으면 백업
    if new_path.exists():
        backup_path = folder / f"old_{new_name}"
        old_path.rename(backup_path)
        print(f"기존 {new_name} → old_{new_name} 로 백업")
    
    old_path.rename(new_path)
    print(f"{old_path.name:40} → {new_name}")

print("\n완료! 이제 이 폴더에 img_001.jpg ~ img_040.jpg 가 깔끔하게 정리되었습니다")
```
